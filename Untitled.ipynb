```python
#Load Libaries
import numpy as np
import pandas as pd
```

```python
#Load Dataset
df = pd.read_csv('data.csv')
```

```python
#To see records
df.shape
```

```python
df.info()
```

```python
#Output - diabetes
#Input - Others 8 columns
```

```python
df['gender'].unique()
```

```python
df['smoking_history'].unique()
```

```python
df['gender'] = df['gender'].map({'Female':0, 'Male':1, 'Other':2})
```

```python
df
```

```python
df['smoking_history'] = df['smoking_history'].map({'never':0, 'No Info':1, 'former':2, 'not current':2, 'ever':3, 'current':4})
```

```python
df
```

```python
#If there are any empty records then delete those records
df.dropna(inplace = True)
```

```python
#To get all columns name
df.columns
```

```python
#Divide dataset into x and y
x = df[['gender', 'age', 'hypertension', 'heart_disease', 'smoking_history','bmi', 'HbA1c_level', 'blood_glucose_level']]
y = df[['diabetes']]
```

```python
#Lets divide into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Lets scale values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train
```

```python
x_train_scaled
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(x_test_scaled)
y_pred = pd.DataFrame(data = y_pred, columns = ['prediction'])
result = pd.concat((y_test.reset_index(drop = True), y_pred), axis = 1)
result
```

```python
#Accuracy of model
from sklearn.metrics import accuracy_score

y_pred = model.predict(x_test_scaled)
y_pred = pd.DataFrame(data = y_pred, columns = ['prediction'])
result = pd.concat((y_test.reset_index(drop = True), y_pred), axis = 1)
score = accuracy_score(result['diabetes'], result['prediction'])
score
```
